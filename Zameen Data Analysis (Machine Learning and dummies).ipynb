### Machine learning

```python
#Create target
target=data_preprocessed['price']
```

```python
target
```

```python
unscaled_inputs=data_preprocessed.drop(['price'],axis=1)
```

```python
unscaled_inputs
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
unscaled_inputs.drop(['latitude','longitude'],axis=1,inplace=True)
```

```python
#Fit the data
scaler.fit(unscaled_inputs)
```

```python
scaled_inputs=scaler.transform(unscaled_inputs)
```

```python
scaled_inputs

```

```python
#Spliting the data into x and y
train_test_split(scaled_inputs,target)
```

```python
x_train,x_test,y_train,y_test=train_test_split(scaled_inputs,target,train_size=0.8,random_state=20)
```

```python
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

```python
#Applying machine learning
linear.fit(x_train,y_train)
```

```python
linear.score(x_train,y_train)
```

```python
#Creating the summary table and see the importance of the each value
coef=linear.coef_
```

```python
intercept=linear.intercept_
```

```python
summary_table=pd.DataFrame(columns=['Features'],data=unscaled_inputs.columns.values)
```
