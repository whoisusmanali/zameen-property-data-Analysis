```python
#Cleaning the null values
raw_data.isnull().sum()
```

```python
agency_nullvalues=raw_data['agency'].isnull().sum()
agent_nullvalues=raw_data['agent'].isnull().sum()
```

```python
raw_data.shape[0]
agency_percentage=(agency_nullvalues/raw_data.shape[0])*100
agent_percentage=(agent_nullvalues/raw_data.shape[0])*100
```

```python
print("Null percentage in Agency",agency_percentage)
print("Null percentage in Agent",agent_percentage)
#The percentage is high so we cannot delete these values 
```

```python
raw_data['agency'].fillna('Self',inplace=True)
raw_data['agent'].fillna('Self',inplace=True)
```

```python
raw_data.head()
```

```python
raw_data.isnull().sum()
#Now data is null free
```

```python
raw_data['Area Type'].unique()
```

```python
#Let's Convert the area of the property from marla and kanal to only marla
#raw_data['Size_in_Marla'].astype('float64')
for i in raw_data.index:
    if raw_data['Area Type'][i]=='Kanal':
        raw_data['Size in Marla'][i]=raw_data['Area Size'][i]*20
    else:
        raw_data['Size in Marla'][i]=raw_data['Area Size'][i]

        
```
